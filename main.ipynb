```python
def preprocess_and_add_user_id(file_path):
    df = pd.read_csv(file_path)

    df.rename(columns=lambda x: x.lower().replace(' ', '_'), inplace=True)
    df.rename(columns={
        'purchase_frequency': 'frequency',
        'purchase_categories': 'categories',
        'add_to_cart_browsing': 'add_to_cart'
    }, inplace=True)

    customer = df.drop(columns=['saveforlater_frequency', 'review_reliability', 'review_helpfulness',
                                'personalized_recommendation_frequency_', 'recommendation_helpfulness',
                                'rating_accuracy_'])

    customer.dropna(inplace=True)

    customer['user_id'] = range(1, len(customer) + 1)

    return customer

file_path = "/Users/juniorlinhares/Documents/projects/Project-IV-sql-tableau/data/Amazon Customer Behaviour.csv"
processed_customer_data = preprocess_and_add_user_id(file_path)
```
